```python
string_columns = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native"]
```

```python
##Preprocessing the data. As, there are categorical variables
number = LabelEncoder()
for i in string_columns:
        train[i]= number.fit_transform(train[i])
        test[i]=number.fit_transform(test[i])
```

```python
##Capturing the column names which can help in futher process
colnames = train.columns
colnames
```

```python
x_train = train[colnames[0:13]]
y_train = train[colnames[13]]
x_test = test[colnames[0:13]]
y_test = test[colnames[13]]
```

```python
##Normalmization
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
x_train = norm_func(x_train)
x_test =  norm_func(x_test)
```

**SVM MODELS**

```python
#kernal=linear
clf1=SVC(kernel='poly',gamma=0.001) # kernel="linear","rbf","polynomial"

clf1.fit(x_train,y_train)

y_pred = clf1.predict(x_test)

acc = accuracy_score(y_test,y_pred)*100
print("Test Accuracy=",acc)

confusion_matrix(y_test, y_pred)
```

```python
    # plot confusion matrix to describe the performance of classifier.
cm_ff=confusion_matrix(y_test, y_pred)
class_label = ["No", "Yes"]
ff_cm = pd.DataFrame(cm_ff, index = class_label, columns = class_label)
plt.figure(figsize=(6, 5))
sns.heatmap(ff_cm, annot = True)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.show()
```

```python
#kernal=linear
clf1=SVC(kernel='linear',gamma=0.01) # kernel="linear","rbf","polynomial"

clf1.fit(x_train,y_train)

y_pred = clf1.predict(x_test)

acc = accuracy_score(y_test,y_pred)*100
print("Test Accuracy=",acc)

confusion_matrix(y_test, y_pred)
```

```python
#kernal=rbf
clf1=SVC(kernel='rbf',gamma=0.001) # kernel="linear","rbf","polynomial"

clf1.fit(x_train,y_train)

y_pred = clf1.predict(x_test)

acc = accuracy_score(y_test,y_pred)*100
print("Test Accuracy=",acc)

confusion_matrix(y_test, y_pred)
```

```python
    # plot confusion matrix to describe the performance of classifier.
cm_ff=confusion_matrix(y_test, y_pred)
class_label = ["No", "Yes"]
ff_cm = pd.DataFrame(cm_ff, index = class_label, columns = class_label)
plt.figure(figsize=(6, 5))
sns.heatmap(ff_cm, annot = True)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.show()
```

```python
#kernal=rbf
clf1=SVC(kernel='rbf',gamma=0.0001,C=1) # kernel="linear","rbf","polynomial"

clf1.fit(x_train,y_train)

y_pred = clf1.predict(x_test)

acc = accuracy_score(y_test,y_pred)*100
print("Test Accuracy=",acc)

confusion_matrix(y_test, y_pred)
```

**From the used model the model with linear kernel and gamma = 0.001 is found to be more accurate to an extent of 80.98%**
